```python
import pandas as pd
df = pd.read_csv('../cohen_data/epc-ir.clean.tsv', header=None, sep='\t')
```

```python
dfj = pd.read_csv('test_cohen.csv')
dfj = dfj[dfj['abstract'] != 'MISSING'].reset_index().drop(['index'], 1)
dfj
```

```python
dfj = dfj.drop(['1', '2', '3','5','6', 'Unnamed: 0'], 1)
```

```python
df = df.rename(columns={2:'study_idx'})
```

```python
dfi = df[df[3] == 'I']
```

```python
dfide = dfi[~dfi.duplicated('study_idx', keep=False)]
dfjde = dfj[~dfj.duplicated('study_idx', keep=False)]
dfjde[['study_idx']] = dfjde[['study_idx']].astype('int64')
```

```python
dfji = dfjde[dfjde['study_idx'].isin(dfide['study_idx'])].reset_index().drop(['index'], 1)
```
