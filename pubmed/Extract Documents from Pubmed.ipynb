```python
from Bio import Entrez
```

```python
from Bio import Medline
from StringIO import StringIO
```

```python
Entrez.email = "successar@gmail.com"
terms = ["Arthroplasty, Replacement, Hip[mesh] Randomized Controlled Trial[ptyp]",
         "Renal Insufficiency[mesh] Randomized Controlled Trial[ptyp]",
         "Respiration, Artificial[MeSH] Infant, Newborn[MeSH] Randomized Controlled Trial[ptyp]"]
```

```python
def search_medline(query, email):
    search = Entrez.esearch(db='pubmed', term=query, usehistory='y')
    handle = Entrez.read(search)
    count = handle['Count']
    search = Entrez.esearch(db='pubmed', term=query, retmax=count)
    handle = Entrez.read(search)
    try:
        return handle
    except Exception as e:
        raise IOError(str(e))
    finally:
        search.close()

def fetch_rec(rec_id, entrez_handle):
    records = {}
    for rec_id in entrez_handle['IdList'] :
        fetch_handle = Entrez.efetch(db='pubmed', id=rec_id,
                                     rettype='Medline', retmode='text')
        rec = fetch_handle.read()
        rec_file = StringIO(rec)
        medline_rec = Medline.read(rec_file)
        if 'AB' in medline_rec :
            records[rec_id] = medline_rec['AB']
        else :
            records[rec_id] = -1
    return records
```

```python
records = []
for search in terms :
    records.append(search_medline(search, ""))
```

```python
rec_0 = fetch_rec(0, records[0])
```

```python
rec_1 = fetch_rec(0, records[1])
```

```python
rec_2 = fetch_rec(0, records[2])
```

```python
rec = [rec_0, rec_1, rec_2]
```

```python
di = {k:(v1, v2) for k,v1,v2 in zip(terms, rec, records)}
```

```python
import cPickle
cPickle.dump(di, open('pubmed_data.p', 'wb'))
```
